# Zscore 

```python
def calculate_z_score(series):
  """Calculates the z-score for a given pandas Series."""
  # Just pass it Pandas Series and it will return a Series of Z Scores
  return (series - series.mean()) / series.std()
```

 ### **General Approach For Each Department** : Get Z Score for Each Feature, then Combine all Individual Feature based Z Score into one by adding up and creating total_z_Score, then Rank by it and get Top N by Sorted Descending on Total Z Score

### A. Batsmen (pick 6)  
Standardise (**z-score**) the following (higher ‚Üí better):  
- `Not Outs`  
- `Runs`  
- `Highest Score`  
- `Batting Average`  
- `Strike Rate`

```python
# 1Ô∏è Define features to use for Z-score

# Calculate z-scores for the selected features individually

# Sum the z-scores for each batsman and add as a new column 'zTot'

# Select the top 6 batsmen based on the total z-score

# Display the selected batsmen's names, country, and total z-score
```

## ANSWERS

```python

batsman_features = ['Not Outs', 'Runs', 'Highest Score', 'Batting Average', 'Strike Rate']


for feature in batsman_features:
    batsman_data[feature + '_z'] = calculate_z_score(batsman_data[feature])


batsman_data['zTotal'] = batsman_data[[f + '_z' for f in batsman_features]].sum(axis=1)


top_6_batsmen = batsman_data.sort_values('zTotal', ascending=False).head(6)


top_6_batsmen[['Player Name', 'Country', 'zTotal']]
```

### B. Bowlers (pick 4)  
Standardise (**z-score**) the following:  
- `Wickets`  (higher better)  
- `Maiden Overs`  (higher better)  
- `Bowling Average`  (lower better)  
- `Economy Rate`  (lower better)

**Alert**

You may need to invert the Value to get Lower Values more Z Score and Higher Value Low Z Score

Bowling Average and Economy Rate are both Negative Metrics which means as Low they are the Good they are. Else you will end up selecting Hasan Ali üòÜ

**How to Invert?**


(1/value) will also make the Inversion with Lower Value being turned into Larger Values

View: https://www.youtube.com/watch?v=cWHQzAEjt5o&authuser=0

```python
# 1Ô∏è‚É£ Invert metrics where lower is better
# Bowling Average and Economy Rate: lower values are better ‚Üí invert
# 2Ô∏è‚É£ Define features to use for Z-score calculation

# 3Ô∏è‚É£ Calculate Z-scores for each feature
# 4Ô∏è‚É£ Sum Z-scores to create a total Z-score

# 5Ô∏è‚É£ Select the top 4 bowlers based on total Z-score

# 6Ô∏è‚É£ Display selected bowlers' name, country, and total Z-score
```

# ANSWERS
##### For bowlers, we need to be extra clever because sometimes less is more (lower averages are better), 
#####  so we have to mathematically "flip" those metrics!

```python

bowler_data['Bowling Average_inv'] = 1 / bowler_data['Bowling Average']
bowler_data['Economy Rate_inv'] = 1 / bowler_data['Economy Rate']


bowler_features = ['Wickets', 'Maiden Overs', 'Bowling Average_inv', 'Economy Rate_inv']


for feature in bowler_features:
    bowler_data[feature + '_z'] = calculate_z_score(bowler_data[feature])


bowler_data['zTot'] = bowler_data[[f + '_z' for f in bowler_features]].sum(axis=1)


top_4_bowlers = bowler_data.sort_values('zTot', ascending=False).head(4)


top_4_bowlers[['Player Name', 'Country', 'zTot']]
```

```python

keeper_features = ['Dismissals', 'Catches', 'Stumpings']


for feature in keeper_features:
    wicketkeeper_data[feature + '_z'] = calculate_z_score(wicketkeeper_data[feature])


wicketkeeper_data['zTot'] = wicketkeeper_data[[f + '_z' for f in keeper_features]].sum(axis=1)


top_keeper = wicketkeeper_data.sort_values('zTot', ascending=False).head(1)


top_keeper[['Player Name', 'Country', 'zTot']]
```
